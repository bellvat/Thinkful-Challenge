## Data Source
UN data for GNI per capita($), Atlas method
http://data.un.org/Data.aspx?d=WDI&f=Indicator_Code%3aNY.GNP.PCAP.CD

UN data for imports of goods and services as a % of GDP
http://data.un.org/Data.aspx?d=WDI&f=Indicator_Code%3aNE.IMP.GNFS.ZS

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
#import csv file containing UN data for imports of goods and services as a % of GDP 2011-2015 for 
#the following 11 countries: Australia, Cambodia, China, India, Israel, Japan, Korea, Mexico, Thailand, US and Vietnam
df_import=pd.read_csv('UNimport.csv', index_col="Country or Area", converters={'Value':float,'Year':int}, usecols=['Country or Area','Year','Value'])
```

```python
df_import
```

```python
#selected 2014 import data
year_import_14 = df_import[(df_import['Year']==2014)].copy()
```

```python
year_import_14
```

```python
#import csv file containing UN data for GNI per Capita ($) 2011-2015 for 
#the following 11 countries: Australia, Cambodia, China, India, Israel, Japan, Korea, Mexico, Thailand, US and Vietnam
df_gni = pd.read_csv('UNgni.csv', 
                     index_col='Country or Area',
                     converters={'Value':int,'Year':int},usecols=['Country or Area','Year','Value'])
```

```python
df_gni
```

```python
#selected 2014 GNI data
year_gni_14= df_gni[(df_gni['Year']==2014)].copy()
```

```python
year_gni_14
```

```python
#added a GNI ($) from 2014 GNI dataframe to 2014 Import dataframe
year_import_14.loc[:,'GNI']=year_gni_14['Value']
```

```python
year_import_14
```

## Visualization 1
Table 1 shows the plotting of data points between % of GDP and GNI per capita for 2014. The scatter plot appears to show a negative correlation between the two data sets for 5 countries with the highest GNI per capita. 

Table 2 shows all countries ranked by GNI per capita for 2014. The 5 countries that appear to have the highest correlation between the data sets include Australia, United States, Japan, Israel and Korea.

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.scatter(x=year_import_14['Value'],
            y=year_import_14['GNI'])
plt.xlabel('% of GDP')
plt.ylabel('GNI per Capita($)')
plt.title('Table 1: 2014 GNI per Capita vs % of GDP')

plt.subplot(1,2,2)
sort_gni=year_import_14['GNI'].sort_values(ascending=False)
x=np.arange(len(year_import_14.index))
countries = sort_gni.index
plt.xticks(x, countries, rotation=90)
plt.bar(x,sort_gni)
plt.title('Table 2: 2014 GNI per Capita by Country')
plt.ylabel('GNI per Capita ($)')

plt.tight_layout()
plt.show()
```

```python
#select import data for years 2011-2014
#2015 data is not present for some countries
year_import_n15=df_import[(df_import['Year']!= 2015)].copy()
```

```python
#select GNI data for years 2011-2014
#2015 data is not present for some countries
year_gni_n15=df_gni[(df_gni['Year']!=2015)].copy()
```

```python
#added a GNI ($) from GNI dataframe to Import dataframe
year_import_n15.loc[:,'GNI']=year_gni_n15['Value']
```

```python
year_import_n15
```

## Visualization 2
Table 3 shows the plotting of data points between % of GDP and GNI per capita for 2011-2014. The scatter plot appears to show a negative correlation between a portion of the data, reinforcing the correlation shown in Visualization 1.

```python
plt.scatter(x=year_import_n15['Value'],y=year_import_n15['GNI'])
plt.xlabel('% of GDP')
plt.ylabel('GNI per Capita($)')
plt.title('Table 3: 2011-2014 GNI per Capita vs % of GDP')
plt.show()
```

## Visualization 3
Table 4 plots imports as a % of GDP across 2011-2014 for the 6 countries with the lowest GNI per Capita, including India, Cambodia, Vietnam, Thailand, China and Mexico. Observations include a trend of increasing imports for Vietnam and Cambodia, and a trend of decreasing imports for Thailand and India. 

```python
india= year_import_n15.loc[year_import_n15.index=='India']
cambodia= year_import_n15.loc[year_import_n15.index=='Cambodia']
vietnam= year_import_n15.loc[year_import_n15.index=='Vietnam']
thailand= year_import_n15.loc[year_import_n15.index=='Thailand']
china= year_import_n15.loc[year_import_n15.index=='China']
mexico= year_import_n15.loc[year_import_n15.index=='Mexico']

plt.plot(india['Year'],india['Value'],color='blue')
plt.plot(cambodia['Year'],cambodia['Value'],color='brown')
plt.plot(vietnam['Year'],vietnam['Value'],color='pink')
plt.plot(thailand['Year'],thailand['Value'],color='green')
plt.plot(china['Year'],china['Value'],color='red')
plt.plot(mexico['Year'],mexico['Value'],color='purple')

blue = mpatches.Patch(color='blue', label='India')
brown = mpatches.Patch(color='brown', label='Cambodia')
pink = mpatches.Patch(color='pink', label='Vietnam')
green = mpatches.Patch(color='green', label='Thailand')
red = mpatches.Patch(color='red', label='China')
purple = mpatches.Patch(color='purple', label='Mexico')
plt.legend(handles=[blue,brown,pink,green,red,purple],bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.ylabel('Imports as a % of GDP')
plt.xlabel('Year')
plt.title('Table 4: Imports as a % of GDP for 6 Countries with Relatively Low GNI: 2011-2014')
plt.show()
```

## Visualization 4
Table 5 plots imports as a % of GDP across 2011-2014 for the 5 countries with relatively high GNI per Capita, including Australia, United States, Japan, Israel and Korea. Observations include a trend of increasing imports for Japan, and a trend of decreasing imports for Korea and Israel. 

```python
aus= year_import_n15.loc[year_import_n15.index=='Australia']
us= year_import_n15.loc[year_import_n15.index=='United States']
jap= year_import_n15.loc[year_import_n15.index=='Japan']
isr= year_import_n15.loc[year_import_n15.index=='Israel']
kor= year_import_n15.loc[year_import_n15.index=='Korea, Rep.']

plt.plot(aus['Year'],aus['Value'],color='blue')
plt.plot(us['Year'],us['Value'],color='brown')
plt.plot(jap['Year'],jap['Value'],color='pink')
plt.plot(isr['Year'],isr['Value'],color='green')
plt.plot(kor['Year'],kor['Value'],color='red')

blue = mpatches.Patch(color='blue', label='Australia')
brown = mpatches.Patch(color='brown', label='United States')
pink = mpatches.Patch(color='pink', label='Japan')
green = mpatches.Patch(color='green', label='Israel')
red = mpatches.Patch(color='red', label='Korea')
plt.legend(handles=[blue,brown,pink,green,red],bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.ylabel('Imports as a % of GDP')
plt.xlabel('Year')
plt.title('Table 5: Imports as a % of GDP for 5 Countries with Relatively High GNI: 2011-2014')
plt.show()
```
